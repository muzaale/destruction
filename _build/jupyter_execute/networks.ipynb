```python
import networkx as nx
import matplotlib.pyplot as plt
#import numpy as np
#import sklearn as skl
# 

plt.figure(figsize=[10, 10])
G = nx.DiGraph()

G.add_node("Teddy Riley",    pos = (.5, 1) )
G.add_node("Babyface",    pos = (1.5, 1) )
G.add_node("Bruno Mars", pos = (.5, 0))
G.add_node("D'Mille", pos = (1.5, 0))

G.add_edge("Teddy Riley", "Bruno Mars")
G.add_edge("Bruno Mars", "Teddy Riley")
G.add_edge("Teddy Riley", "Babyface")
G.add_edge("Babyface", "D'Mille")
G.add_edge("D'Mille", "Babyface")


#G.add_edge("Local", "GH")
nx.draw(G, 
        nx.get_node_attributes(G, 'pos'), 
        with_labels=True, 
        font_weight='bold', 
        node_size = 2500,
        node_color = "lightblue",
        linewidths = 3)
ax= plt.gca()
ax.collections[0].set_edgecolor("#000000")
ax.set_xlim([-.3, 2.3])
ax.set_ylim([-.3, 1.3])
plt.show()
```
